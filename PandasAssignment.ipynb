````python
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
````

````python
df = pd.read_csv("travel-times.csv")
df.shape
````

````python
df.head(15)
````

````python
df.info()
````

````python
df_wo_duplicates = df.drop_duplicates(inplace=False)
df_wo_duplicates
````

````python
missing_values = df.isnull().sum().sum()
print("Total number of missing values:", missing_values)
````

````python
df_copyA= df.copy()

df_missing_rows = df_copyA.dropna()
df_missing_rows
````

````python
df_copyB= df.copy()

df_missing_columns = df_copyB.dropna(axis=1)
df_missing_columns
````

````python
summary = df.describe()
summary
````

````python
subset = df[df['Date'].isin(['11/23/2011','1/6/2012'])]
subset
````

````python
df.plot(kind='scatter', x= 'Distance', y='TotalTime', title='Distance vs Total Time');
````

````python
df.boxplot(column='AvgSpeed')
plt.title('Average Speed Boxplot')
plt.ylabel('Average Speed')
plt.show()
````

````python
df.boxplot(column='AvgMovingSpeed')
plt.title('Average Moving Speed Boxplot')
plt.ylabel('Average Moving Speed')
plt.show()
````
